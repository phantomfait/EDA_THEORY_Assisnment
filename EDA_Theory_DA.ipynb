```python
import pandas as pd

file_path = 'melb_data.csv'
data = pd.read_csv(file_path)


data_info = {
    'Dimensions': data.shape,
    'Columns': data.columns.to_list(),
    'Data Types': data.dtypes,
    'Missing Values': data.isnull().sum(),
    'Data Summary': data.describe(include='all')
}


for key, value in data_info.items():
    print(f"\n{key}:\n", value)


data['Car'].fillna(data['Car'].median(), inplace=True)


data['BuildingArea'].fillna(data['BuildingArea'].median(), inplace=True)


data['YearBuilt'].fillna(data['YearBuilt'].median(), inplace=True)


data['CouncilArea'].fillna(data['CouncilArea'].mode()[0], inplace=True)


missing_values_after_cleaning = data.isnull().sum()
print("\nMissing Values After Cleaning:\n", missing_values_after_cleaning)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Univariate Analysis (Numerical features: mean, median, mode, std, etc.)
numerical_features = ['Price', 'Rooms', 'Distance', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'YearBuilt']
for feature in numerical_features:
    print(f'\n--- {feature} ---')
    print(f'Mean: {data[feature].mean()}')
    print(f'Median: {data[feature].median()}')
    print(f'Mode: {data[feature].mode()[0]}')
    print(f'Standard Deviation: {data[feature].std()}')
    print(f'Min: {data[feature].min()}')
    print(f'Max: {data[feature].max()}')
    print(f'25th Percentile: {data[feature].quantile(0.25)}')
    print(f'75th Percentile: {data[feature].quantile(0.75)}')

data[numerical_features].hist(bins=30, figsize=(15, 10), layout=(3, 3))
plt.tight_layout()
plt.show()

categorical_features = ['Suburb', 'Type', 'Method', 'SellerG', 'CouncilArea', 'Regionname']
for feature in categorical_features:
    print(f'\n--- {feature} ---')
    print(f'Mode: {data[feature].mode()[0]}')
    print(f'Unique Values: {data[feature].nunique()}')
    print(f'Value Counts:\n{data[feature].value_counts().head()}')

    plt.figure(figsize=(10, 4))
    sns.countplot(data[feature], order=data[feature].value_counts().index)
    plt.xticks(rotation=90)
    plt.show()

# Bivariate Analysis
for feature in numerical_features:
    if feature != 'Price':
        print(f'\n--- Price vs {feature} ---')
        print(data[['Price', feature]].corr())

        plt.figure(figsize=(6, 4))
        sns.scatterplot(x=data[feature].reset_index(drop=True), y=data['Price'].reset_index(drop=True))
        plt.show()

for feature in categorical_features:
    plt.figure(figsize=(10, 4))
    sns.boxplot(x=data[feature].reset_index(drop=True), y=data['Price'].reset_index(drop=True))
    plt.xticks(rotation=90)
    plt.show()

correlation_matrix = data[numerical_features].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.show()

# Multivariate Analysis
sns.pairplot(data[numerical_features].reset_index(drop=True))
plt.show()

plt.figure(figsize=(12, 8))
sns.heatmap(data[numerical_features].corr(), annot=True, cmap='YlGnBu', linewidths=0.5)
plt.show()
```

```python


# Contingency table for two categorical variables 'Type' and 'Regionname'
contingency_table_type_region = pd.crosstab(data['Type'], data['Regionname'])
print("Contingency Table between Property Type and Region Name:")
print(contingency_table_type_region)

# Another for 'Type' and 'CouncilArea'
contingency_table_type_council = pd.crosstab(data['Type'], data['CouncilArea'])
print("\nContingency Table between Property Type and Council Area:")
print(contingency_table_type_council)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.spatial import ConvexHull
import seaborn as sns


np.random.seed(123)




data_subset = data[['Price', 'Distance']].dropna()


data_subset = data_subset.sample(n=200, random_state=123)


k = 3
kmeans = KMeans(n_clusters=k, n_init=25, random_state=123)
data_subset['cluster'] = kmeans.fit_predict(data_subset)


plt.figure(figsize=(10, 6))
colors = sns.color_palette("hsv", k)

for i in range(k):

    cluster_points = data_subset[data_subset['cluster'] == i][['Distance', 'Price']].values
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], s=30, color=colors[i], label=f'Cluster {i+1}')


    if len(cluster_points) >= 3:
        hull = ConvexHull(cluster_points)
        for simplex in hull.simplices:
            plt.plot(cluster_points[simplex, 0], cluster_points[simplex, 1], color=colors[i], alpha=0.3)


plt.title("Clusters of Properties based on Price and Distance")
plt.xlabel("Distance from CBD")
plt.ylabel("Price (AUD)")
plt.legend()
plt.grid(True)
plt.show()
```

```python

from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import pdist
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler




numeric_data = data[['Rooms', 'Price', 'Distance', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Lattitude', 'Longtitude']].dropna()


numeric_data_sample = numeric_data.sample(n=100, random_state=123)


scaler = StandardScaler()
scaled_data_sample = scaler.fit_transform(numeric_data_sample)


def plot_dendrogram(linkage_matrix, title):
    plt.figure(figsize=(10, 6))
    dendrogram(linkage_matrix)
    plt.title(title)
    plt.xlabel("Data Points")
    plt.ylabel("Distance")
    plt.show()


euclidean_dist = pdist(scaled_data_sample, metric="euclidean")
euclidean_linkage = linkage(euclidean_dist, method="ward")
plot_dendrogram(euclidean_linkage, "Dendrogram (Euclidean Distance) 22BCE0341")


manhattan_dist = pdist(scaled_data_sample, metric="cityblock")
manhattan_linkage = linkage(manhattan_dist, method="ward")
plot_dendrogram(manhattan_linkage, "Dendrogram (Manhattan Distance) 22BCE0341")


cosine_dist = pdist(scaled_data_sample, metric="cosine")
cosine_linkage = linkage(cosine_dist, method="ward")
plot_dendrogram(cosine_linkage, "Dendrogram (Cosine Distance) 22BCE0341")
```

```python

from sklearn.decomposition import PCA


numeric_data = data[['Rooms', 'Price', 'Distance', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Lattitude', 'Longtitude']].dropna()


scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_data)


pca = PCA()
pca_result = pca.fit_transform(scaled_data)


explained_variance = pca.explained_variance_ratio_
print("Explained Variance Ratio by Each Principal Component:", explained_variance)


plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.title('Scree Plot: Variance Explained by Each Principal Component 22BCE0341')
plt.xticks(range(1, len(explained_variance) + 1))
plt.grid()
plt.show()


np.random.seed(123)
sample_indices = np.random.choice(range(len(pca_result)), size=500, replace=False)
sampled_pca_data = pca_result[sample_indices]


plt.figure(figsize=(10, 8))
plt.scatter(sampled_pca_data[:, 0], sampled_pca_data[:, 1], color='red', alpha=0.5, label='Data Points')
for i, vector in enumerate(pca.components_):
    plt.arrow(0, 0, vector[0] * max(sampled_pca_data[:, 0]), vector[1] * max(sampled_pca_data[:, 1]),
              color='blue', alpha=0.8, linewidth=2)
    plt.text(vector[0] * max(sampled_pca_data[:, 0]) * 1.15,
             vector[1] * max(sampled_pca_data[:, 1]) * 1.15,
             data.columns[i], color='blue', ha='center', va='center', fontsize=12)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Biplot with Sampled Data Points 22BCE0341")
plt.grid()
plt.legend()
plt.show()
```
